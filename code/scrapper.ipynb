Scrape an url - normal beautiful soup

```python
import requests
from bs4 import BeautifulSoup

def scrape(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    return soup
```

```python
url = "https://www.uni-bamberg.de/its/"
```

```python
data = scrape(url)
```

```python
print(data)
```
